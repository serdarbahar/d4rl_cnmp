```python
import numpy as np
import pickle

demos = pickle.load(open('relocate.pickle', 'rb'))

for path in demos:
    path['init_state_dict'] = path['init_state_dict'].copy()
    print(path['init_state_dict']['qpos'])
    print(path['init_state_dict']['qvel'])
    print(path['init_state_dict']['obj_pos'])
    print(path['init_state_dict']['target_pos'])
    print(path['init_state_dict']['palm_pos'])

    P_O = -1*path['init_state_dict']['obj_pos'] + path['init_state_dict']['palm_pos']
    P_T = -1*path['init_state_dict']['target_pos'] + path['init_state_dict']['palm_pos']
    O_T = path['init_state_dict']['obj_pos'] - path['init_state_dict']['target_pos']

    print(f"Palm - Object: {P_O}")
    print(f"Palm - Target: {P_T}")
    print(f"Object - Target: {O_T}")
    

    print()
    
```
